# Nutzung des lokalen Modells Llama

### Nutzung eines lokal gehosteten LLM über die OpenAI-kompatible API. Der Server läuft auf `http://localhost:8000/v1`.

```python
from openai import OpenAI

client = OpenAI(base_url="http://localhost:8000/v1", api_key="none")


response = client.chat.completions.create(      
    model="meta-llama/Llama-3.1-8B-Instruct",
    messages=[{"role": "user", "content": "welches modell bist du ? kurz"}]
)

print(response.choices[0].message.content)
```

### Nutzung des Modells für Textgenerierung mit Langchain

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="meta-llama/Llama-3.1-8B-Instruct",
    base_url="http://localhost:8000/v1", 
    api_key="none",                       
    temperature=1                         
)

# Einfache Anfrage
msg = llm.invoke("welches modell bist du ? kurz")


print(msg.content)
```

### Nutzung des Modells für Embedding analog mit openai und langchain. Der Server läuft auf `http://localhost:8002/v1`.

```python
from openai import OpenAI

client = OpenAI(base_url="http://localhost:8002/v1", api_key="none")


response = client.chat.completions.create(      
    model="meta-llama/Llama-3.1-8B-Instruct",
    messages=[{"role": "user", "content": "welches modell bist du ? kurz"}]
)

#print(response.choices[0].message.content)
```

```python
response = client.embeddings.create(
    model="meta-llama/Llama-3.1-8B-Instruct",
    input="hallo, das ist nur ein test"
)
```

```python
print(response.data[0].embedding)
```

### Mit langchain

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    model="meta-llama/Llama-3.1-8B-Instruct",   
    base_url="http://localhost:8002/v1",        
    api_key="none"                              
)

# Einzelner Text
vec = embeddings.embed_query("Hallo Welt")
print(len(vec))  

# Mehrere Texte
vecs = embeddings.embed_documents([
    "Berlin ist die Hauptstadt von Deutschland.",
    "München hat das Oktoberfest."
])
print(len(vecs), len(vecs[0]))
```
