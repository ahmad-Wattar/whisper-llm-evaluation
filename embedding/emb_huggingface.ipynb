## Model google/embeddinggemma-300m (multilingual)

```python
from huggingface_hub import InferenceClient
import numpy as np

# Initialisiere Client mit Embedding-Modell
client = InferenceClient(model="google/embeddinggemma-300m", token="your token")

query = "Which planet is known as the Red Planet?"
documents = [
    "Venus is often called Earth's twin because of its similar size and proximity.",
    "Mars, known for its reddish appearance, is often referred to as the Red Planet.",
    "Jupiter, the largest planet in our solar system, has a prominent red spot.",
    "Saturn, famous for its rings, is sometimes mistaken for the Red Planet."
]

# Embeddings abrufen
query_emb = client.feature_extraction(query)
doc_embs = [client.feature_extraction(doc) for doc in documents]
```

```python
# Embeddings ausgeben
print("=== Query Embedding ===")
print(query_emb)
print("\nShape:", query_emb.shape)

print("\n=== Document Embeddings ===")
for i, emb in enumerate(doc_embs):
    print(f"Document {i+1}: {documents[i]}")
    print(emb)
    print("Shape:", emb.shape)
    print()
```

## Kosinusähnlichkeit der verschiedenen Dokumente im Vergleich zur Query

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np


#  sklearn für die Similarity
similarities = cosine_similarity(query_emb.reshape(1, -1), doc_embs)[0]

# Ranking
ranking = np.argsort(similarities)[::-1]
print("\n=== Similarities ===")
for i in ranking:
    print(f"Similarity: {similarities[i]:.4f} | Document: {documents[i]}")
```
