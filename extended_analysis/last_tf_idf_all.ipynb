### Clean text

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\d+", " ", text)    # Zahlen löschen
    text = re.sub(r"[^\wäöüß ]", " ", text)  # Sonderzeichen
    text = re.sub(r"\s+", " ", text)    # Mehrfach-Leerzeichen
    return text

df2["clean"] = df2["text"].apply(clean_text)
df1["clean"] = df1["text"].apply(clean_text)
```

### Stopwords löschen

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import nltk

# Lade deutsche Stopwords herunter
nltk.download("stopwords")

# Lade deutsche Stopwords
german_stopwords = stopwords.words("german")

#  Die bisherigen zusätzlichen Stopwords
custom_stopwords = german_stopwords + [
    "ja", "mal", "nein", "okay", "gut", "also", "genau",
    "äh", "ähm", "hm", "hmm", "super", "bitte", "danke",
    "oder", "halt", "eigentlich", "wirklich", "vielleicht",
    "eben", "naja"
]

# ==> Neue, extrahierte Gesprächswörter aus TF-IDF-Verlustanalyse
dialog_stopwords = [
    "aha",
    "bisschen",
    "deswegen",
    "doktor",
    "frau",
    "fragen",
    "ganz",
    "gerne",
    "gesagt",
    "gibt",
    "herr",
    "herr doktor",
    "immer",
    "klar",
    "mehr",
    "nehmen",
    "paar",
    "sagen",
    "schon",
    "sieht",
    "tag",
    "wann",
    "weiß",
    "wissen",
    "wunderbar",
    "zeit",
    "schön",
    "nee",
    "ne",
    "heißt",
    "irgendwelche",
    "irgendwelcher",
    "geht",
    "dank"
]

# Zusammenführen + deduplizieren
all_stopwords = list(set(custom_stopwords + dialog_stopwords))

print("Gesamtanzahl Stopwords:", len(all_stopwords))
```

## delta-Score Berechnung

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# --- Modelle, die NICHT analysiert werden sollen ---
skip_models = ["GPT-5", "OpenAI GPT-5", "gpt-5", "openai gpt-5"]


def compute_mean_tfidf_results(df1, df2, top_k=20):

    results = []

    for i, row in df2.iterrows():

        audio_id = row["audio_id"]
        transkript_text = row["clean"]

        print(f"🔍 Processing audio_id = {audio_id}")

        df_sums = df1[df1["audio_id"] == audio_id].copy()

        if df_sums.empty:
            print(f"⚠️ Keine Summaries für {audio_id} gefunden — überspringe.")
            continue

        # ---------------------------------------
        # 0. Modelle herausfiltern (GPT-5 usw.)
        # ---------------------------------------
        df_sums = df_sums[~df_sums["model_name"].isin(skip_models)]

        if df_sums.empty:
            print(f"⚠️ Alle Summaries für {audio_id} gehören zu skip-Models. Überspringe.")
            continue

        # ---------------------------------------
        # 1. TF-IDF nur auf Transkript fitten
        # ---------------------------------------
        tfidf = TfidfVectorizer(
            stop_words=all_stopwords,
            ngram_range=(1,2),
            max_features=None
        )

        X_trans = tfidf.fit_transform([transkript_text]).toarray()[0]
        feature_names = tfidf.get_feature_names_out()

        # ---------------------------------------
        # 2. Summaries transformieren
        # ---------------------------------------
        X_sums = tfidf.transform(df_sums["clean"]).toarray()
        df_sums["tfidf_vector"] = list(X_sums)

        # ---------------------------------------
        # 3. Gruppieren nach Modell (ohne GPT-5)
        # ---------------------------------------
        for model_name, group in df_sums.groupby("model_name"):

            summary_matrix = np.stack(group["tfidf_vector"].values)
            mean_vec = summary_matrix.mean(axis=0)

            delta = X_trans - mean_vec

            top_idx = np.argsort(-delta)[:top_k]
            lost_keywords = [(feature_names[j], float(delta[j])) for j in top_idx]

            results.append({
                "audio_id": audio_id,
                "model_name": model_name,
                "lost_keywords": lost_keywords,
                "delta_vector": delta.tolist(),
                "mean_summary_vector": mean_vec.tolist()
            })

    return pd.DataFrame(results)

# --------------------------------------------------
# PIPELINE AUSFÜHREN
# --------------------------------------------------
result_df = compute_mean_tfidf_results(df1, df2, top_k=20)

result_df.to_csv("tfidf_mean_results_filtered.csv", index=False)

print("✅ Fertig! Ergebnisse gespeichert (GPT-5 Modelle wurden übersprungen).")
```

```python
import numpy as np

result_df["delta_vector"] = result_df["delta_vector"].apply(np.array)


result_df["delta_score"] = result_df["delta_vector"].apply(lambda v: np.mean(np.abs(v)))
```

```python
model_scores = (
    result_df
    .groupby("model_name")["delta_score"]
    .agg(
        mean="mean",
        min="min",
        max="max",
        std="std"
    )
    .reset_index()
    .sort_values("mean")   # kleiner = besser
)
```

# Delta und  Kosinus zusammen berechnen 

```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# --------------------------------------------------
# MODELLE, DIE NICHT ANALYSIERT WERDEN SOLLEN
# --------------------------------------------------
skip_models = ["GPT-5", "OpenAI GPT-5", "gpt-5", "openai gpt-5"]


# --------------------------------------------------
# TF-IDF + MEAN-SUMMARY PIPELINE (OHNE WORTANALYSE)
# --------------------------------------------------
def compute_mean_tfidf_results(df1, df2, all_stopwords):

    results = []

    for _, row in df2.iterrows():

        audio_id = row["audio_id"]
        transcript_text = row["clean"]

        print(f"🔍 Processing audio_id = {audio_id}")

        df_sums = df1[df1["audio_id"] == audio_id].copy()

        if df_sums.empty:
            continue

        # Skip-Modelle entfernen
        df_sums = df_sums[~df_sums["model_name"].isin(skip_models)]
        if df_sums.empty:
            continue

        # TF-IDF nur auf Transkript fitten
        tfidf = TfidfVectorizer(
            stop_words=all_stopwords,
            ngram_range=(1, 2),
            max_features=5000
        )

        X_trans = tfidf.fit_transform([transcript_text]).toarray()[0]

        # Summaries transformieren
        X_sums = tfidf.transform(df_sums["clean"]).toarray()
        df_sums["tfidf_vector"] = list(X_sums)

        # Gruppieren nach Modell
        for model_name, group in df_sums.groupby("model_name"):

            summary_matrix = np.stack(group["tfidf_vector"].values)
            mean_vec = summary_matrix.mean(axis=0)

            delta_vec = X_trans - mean_vec
            delta_score = np.mean(np.abs(delta_vec))

            cos_score = cosine_similarity(
                X_trans.reshape(1, -1),
                mean_vec.reshape(1, -1)
            )[0, 0]

            results.append({
                "audio_id": audio_id,
                "model_name": model_name,
                "delta_score": delta_score,
                "cosine_similarity": cos_score
            })

    return pd.DataFrame(results)


# --------------------------------------------------
# PIPELINE AUSFÜHREN
# --------------------------------------------------
result_df = compute_mean_tfidf_results(
    df1=df1,
    df2=df2,
    all_stopwords=all_stopwords
)

print("\n================ EINZELERGEBNISSE =================")
print(result_df)


# --------------------------------------------------
# AGGREGATION: DELTA-SCORE
# --------------------------------------------------
delta_stats = (
    result_df
    .groupby("model_name")["delta_score"]
    .agg(
        mean="mean",
        min="min",
        max="max",
        std="std"
    )
    .reset_index()
    .sort_values("mean")   # kleiner = besser
)

print("\n================ DELTA SCORE (kleiner = besser) =================")
print(delta_stats)


# --------------------------------------------------
# AGGREGATION: COSINUS-SIMILARITY
# --------------------------------------------------
cosine_stats = (
    result_df
    .groupby("model_name")["cosine_similarity"]
    .agg(
        mean="mean",
        min="min",
        max="max",
        std="std"
    )
    .reset_index()
    .sort_values("mean", ascending=False)  # größer = besser
)

print("\n================ COSINUS SCORE (größer = besser) =================")
print(cosine_stats)
```

```python
import pandas as pd

model_stability = (
    result_df.groupby("model_name")["delta_score"]
    .std()
    .reset_index()
    .rename(columns={"delta_score": "delta_std"})
    .sort_values("delta_std")
)

print(model_stability)
```

## Score für einzelne Zusammenfassungen

```python
# ----------------------------------------
# NEU: Delta Score pro einzelner Summary
# ----------------------------------------

single_results = []

for i, row in df2.iterrows():

    audio_id = row["audio_id"]
    transcript_text = row["clean"]

    # TF-IDF nur auf Transkript fitten
    tfidf = TfidfVectorizer(
        stop_words=all_stopwords,
        ngram_range=(1,2),
        max_features=None
    )
    X_trans = tfidf.fit_transform([transcript_text]).toarray()[0]

    # Alle Summaries zu diesem Transkript
    df_sums = df1[df1["audio_id"] == audio_id].copy()
    df_sums = df_sums[~df_sums["model_name"].isin(skip_models)]

    if df_sums.empty:
        continue

    # Summaries transformieren
    X_sums = tfidf.transform(df_sums["clean"]).toarray()

    # Jetzt pro einzelne Summary delta berechnen
    for (idx, row_sum), vector in zip(df_sums.iterrows(), X_sums):

        delta_vec = X_trans - vector
        delta_score_single = float(np.mean(np.abs(delta_vec)))

        single_results.append({
            "audio_id": audio_id,
            "model_name": row_sum["model_name"],
            "run_id": row_sum["run_id"],
          
            "delta_score_single": delta_score_single
        })

df_single = pd.DataFrame(single_results)
```

## Bag-of-Words Frequenzen für Transkripte (df2)

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

vectorizer = CountVectorizer(stop_words=custom_stopwords)

# Frequenzen für alle Transkripte
X_transcripts = vectorizer.fit_transform(df2["text"])
vocab = vectorizer.get_feature_names_out()

df2_freq = pd.DataFrame(X_transcripts.toarray(), columns=vocab)
df2_freq["audio_id"] = df2["audio_id"]

df2_freq.head(5)
#print(df2_freq.shape[1])


```

# Bag-of-Words Frequenzen für Summaries (df1)

```python
# Gleicher Vectorizer wie oben
X_summaries = vectorizer.transform(df1["text"])

df1_freq = pd.DataFrame(X_summaries.toarray(), columns=vocab)
df1_freq["audio_id"] = df1["audio_id"]
df1_freq["model_name"] = df1["model_name"]
df1_freq["run_id"] = df1["run_id"]

df1_freq.head()
```

# Vergleich pro audio_id: Frequenz Original vs. Summary

```python
def compare_word_freq(summary_row):
    audio_id = summary_row["audio_id"]

    # Original-Frequenzen (eine Zeile)
    orig_vec = df2_freq[df2_freq["audio_id"] == audio_id].iloc[0][vocab].values

    # Summary-Frequenzen (eine Zeile)
    sum_vec = summary_row[vocab].values

    diff = sum_vec - orig_vec  # positive → Überbetont im Summary
                               # negative → Unterrepräsentiert im Summary

    return diff

# Neue Matrix: Differenzen pro Summary
freq_diff = df1_freq.apply(compare_word_freq, axis=1)
```

```python
skip_models = ["OpenAI GPT-5", "GPT-5"]

freq_diff_df = pd.DataFrame(freq_diff.tolist(), columns=vocab)

freq_diff_df["model_name"] = df1_freq["model_name"].values
freq_diff_df["audio_id"] = df1_freq["audio_id"].values

# Informationsverlust pro Summary
freq_diff_df["info_loss"] = (
    -freq_diff_df[vocab]
    .where(freq_diff_df[vocab] < 0, 0)
    .sum(axis=1)
)

# Nur mean pro Modell + Skip-Modelle
model_means = (
    freq_diff_df
    .loc[~freq_diff_df["model_name"].isin(skip_models)]
    .groupby("model_name")["info_loss"]
    .mean()
    .sort_values()
)

print(model_means)
```

# Missing Words 

```python
# --------------------------------------------------
# MODELLE, DIE ÜBERSPRUNGEN WERDEN SOLLEN
# --------------------------------------------------
skip_models = ["OpenAI GPT-5", "GPT-5"]


# --------------------------------------------------
# FEHLENDE WÖRTER PRO SUMMARY
# --------------------------------------------------
def missing_words(summary_row):
    audio_id = summary_row["audio_id"]

    # Wörter im Original (binär)
    orig_words = set(
        df2_freq[df2_freq["audio_id"] == audio_id][vocab]
        .astype(bool)
        .iloc[0]
        .pipe(lambda x: x.index[x.values])
    )

    # Wörter im Summary (binär)
    sum_words = set(
        summary_row[vocab]
        .astype(bool)
        .pipe(lambda x: x.index[x.values])
    )

    # Wörter, die im Summary fehlen
    return list(orig_words - sum_words)


# --------------------------------------------------
# BERECHNUNG
# --------------------------------------------------
df1_freq["missing_words"] = df1_freq.apply(missing_words, axis=1)
df1_freq["missing_word_count"] = df1_freq["missing_words"].apply(len)


# --------------------------------------------------
# AGGREGATION NACH MODELL (MEAN)
# --------------------------------------------------
model_missing_means = (
    df1_freq
    .loc[~df1_freq["model_name"].isin(skip_models)]
    .groupby("model_name")["missing_word_count"]
    .mean()
    .sort_values()
)

print(model_missing_means)
```
